```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
```

```python
np.random.seed(5)
iris = datasets.load_iris()
X = iris.data
np.random.seed(200)
k = 3
# centroids[i] = [x, y]
centroids = {
    i: X[i,:].tolist()
    for i in range(k)
}
print("Initial Centroids:")
print(centroids)
```

```python
df=pd.DataFrame(X)
df.columns=['sepal length','sepal width','petal length','petal width']
```

```python
#Assignment Step
def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1 - x2)^2 - (y1 - y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['sepal length'] - centroids[i][0]) ** 2
                + (df['sepal width'] - centroids[i][1]) ** 2
                +(df['petal length'] - centroids[i][2]) ** 2
                +(df['petal width'] - centroids[i][3]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    #print(centroid_distance_cols)
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    #print(df['closest'])
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    #print(df['closest'])
    return df
```

```python
## Update Stage
import copy
def update(k):
    old_centroids = copy.deepcopy(k)
    print(old_centroids)
    for i in k.keys():
        k[i][0] = np.mean(df[df['closest'] == i]['sepal length'])
        k[i][1] = np.mean(df[df['closest'] == i]['sepal width'])
        k[i][2] = np.mean(df[df['closest'] == i]['petal length'])
        k[i][3] = np.mean(df[df['closest'] == i]['petal width'])
    return k
#centroids = update(centroids)
print("Centroids after first iteration:")
print(centroids)
```

```python
count=2
df1 = assignment(df, centroids)
while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = assignment(df, centroids)
    print("After {} iteration centroids are:".format(count))
    print(centroids)
    count= count+1
    if closest_centroids.equals(df['closest']):
        break
```
