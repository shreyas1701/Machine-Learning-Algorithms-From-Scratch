```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('C:/Users/shrey/Downloads/ML/SCLC_study_output_filtered_2.csv',header=None)
df1
```

```python
df1=df1.drop(df1.index[0])
df1 = df1.reset_index(drop=True)
df1
```

```python
X = df1.iloc[:,1:]
for i in range(0,40):
    if(i<=19):
        df1.iloc[i,0]=1
    else:
        df1.iloc[i,0]=2
```

```python
y1=df1.iloc[:,0]
```

```python
X =np.asarray(X)
y1=np.asarray(y1)
```

```python
mean_vectors = []
for cl in range(1,3):
    mean_vectors.append(np.mean(X[y1==cl], axis=0))
    print('Mean Vector class %s: %s\n' %(cl, mean_vectors[cl-1]))
```

```python
S_W = np.zeros((19,19))
for cl,mv in zip(range(1,3), mean_vectors):
    class_sc_mat = np.zeros((19,19))                  # scatter matrix for every class
    for each in X[y1 == cl]:
        each, mv = each.reshape(19,1), mv.reshape(19,1) # make column vectors
        class_sc_mat += (each-mv).dot((each-mv).T)
    S_W += class_sc_mat                             # sum class scatter matrices
print('within-class Scatter Matrix:\n', S_W)
```

```python
S_W.shape
```

```python
overall_mean = np.mean(X, axis=0)
S_B = np.zeros((19,19))
for i,mean_vec in enumerate(mean_vectors):  
    n = X[y1==i+1,:].shape[0]
    mean_vec = mean_vec.reshape(19,1) # make column vector
    overall_mean = overall_mean.reshape(19,1) # make column vector
    S_B += n * (mean_vec - overall_mean).dot((mean_vec - overall_mean).T)

print('between-class Scatter Matrix:\n', S_B)

eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
for i in range(len(eig_vals)):
    eigvec_sc = eig_vecs[:,i].reshape(19,1)   
    print('\nEigenvector {}: \n{}'.format(i+1, eigvec_sc.real))
    print('Eigenvalue {:}: {:.2e}'.format(i+1, eig_vals[i].real))

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in decreasing order:\n')
for i in eig_pairs:
    print(i[0])

print('Variance explained:\n')
eigv_sum = sum(eig_vals)
for i,j in enumerate(eig_pairs):
    print('eigenvalue {0:}: {1:.2%}'.format(i+1, (j[0]/eigv_sum).real))
```

```python
W = np.hstack(eig_pairs[0][1].reshape(19,1))
print('Matrix W:\n', W.real)
```

```python
X_lda = X.dot(W)
X_lda=X_lda.reshape((40,1))
```

```python
#plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_title('Apply LDA to a toy dataset')
ax.plot(-X_lda[0:20], np.zeros(20), linestyle='None', marker='o', color='blue', label='NSCLC')
ax.plot(-X_lda[20:40], np.zeros(20), linestyle='None', marker='o', color='red', label='SCLC')
fig.show()
```

## LDA using sklearn library


```python
#Data for Sklearn LDA
X_dash =df1.iloc[:,1:]
y1_dash=df1.iloc[:,0]
```

```python
y1_dash =np.array(y1_dash,dtype=int)
```

```python
y1_dash
```

```python
# LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
sklearn_lda = LDA(n_components=1,solver = 'eigen')
X_lda_sklearn = sklearn_lda.fit_transform(X_dash, y1_dash)
```

```python
X_lda_sklearn= -X_lda_sklearn
```

```python
#plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_title('Apply LDA to a toy dataset')
ax.plot(X_lda_sklearn[0:20], np.zeros(20), linestyle='None', marker='o', color='blue', label='NSCLC')
ax.plot(X_lda_sklearn[20:40], np.zeros(20), linestyle='None', marker='o', color='red', label='SCLC')
fig.show()
```
